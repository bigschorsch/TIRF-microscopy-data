```python
import matplotlib.pyplot as plt
import numpy as np
import os

save_dir = r"/Users/Schorschi/Desktop/Masterarbeit/LAB-Data/Georg/37vs26degrees"

fname = "roi_intensity_change_over_time_26_vs_37degrees_L1"
# Werte aus deinen Fits
slopes = {
    "26°C (2 SLBs)": {
        "m": -0.00032,
        "std": 0.000298
    },
    "37°C (2 SLBs)": {
        "m": -0.00064,
        "std": 0.001118
    }
}

# Für einen Boxplot brauchen wir Stichproben.
# Wir simulieren je 100 Punkte aus N(m, std)
np.random.seed(0)
data_24 = np.random.normal(slopes["26°C (2 SLBs)"]["m"], slopes["26°C (2 SLBs)"]["std"], 100)
data_37 = np.random.normal(slopes["37°C (2 SLBs)"]["m"], slopes["37°C (2 SLBs)"]["std"], 100)

# Boxplot
plt.figure(figsize=(6, 4))
plt.boxplot([data_24, data_37], labels=["26°C (2 SLBs)", "37°C (2 SLBs)"])
plt.ylabel("Trend slope m [1/min]")
plt.title("Comparison of trend slopes (ROI Intensity change)")
plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.tight_layout()
plt.show()
plt.savefig(os.path.join(save_dir, f'{fname}.png'), dpi=300)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import os

save_dir = r"/Users/Schorschi/Desktop/Masterarbeit/LAB-Data/Georg/37vs26degrees"

fname = "forest_plot_roi_intensity_change_over_time_26_vs_37degrees_pmhc"

labels = ["26°C (4 SLBs)", "37°C (4 SLBs)"]
means = np.array([-0.00123, -0.00049])
stds = np.array([0.0005, 0.001028])   # besser CI, falls n bekannt

y = np.arange(len(labels))

plt.figure(figsize=(6,3))
plt.errorbar(means, y, xerr=stds, fmt='o', capsize=5)
plt.axvline(0, color='gray', linestyle='--')
plt.yticks(y, labels)
plt.xlabel("Trend slope m [1/min]")
plt.title("Forest plot of ROI intensity change")
plt.tight_layout()
plt.savefig(os.path.join(save_dir, f'{fname}.png'))
plt.show()

```
